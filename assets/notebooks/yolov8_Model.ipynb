```python
# This is a conceptual example for model loading
from ultralytics import YOLO

# Load a smaller, pre-trained YOLOv8 model
model = YOLO('yolov8s.pt') # 'yolov8n.pt' for nano, 'yolov8s.pt' for small
```

```python
import os

# List the contents of the downloaded dataset directory
print(os.listdir('Annotate_Faces-2'))

```

```python
# Train on your custom dataset
model.train(
    data="Annotate_Faces-2/data.yaml",  # correct dataset path
    epochs=50,
    imgsz=512,
    batch=16
)
```
